```python
semisupervised_train_for_model=semisupervised_train.drop(['id'], axis=1)
semisupervised_train_for_model
```

```python
semisupervised_train_for_model = pd.get_dummies(semisupervised_train_for_model)
semisupervised_train_for_model
```

```python
from sklearn.model_selection import train_test_split
train, validation = train_test_split(semisupervised_train_for_model, test_size=0.1, random_state=42)
```

```python
train = pd.concat([train['recommend'], train.drop(['recommend'], axis=1)], axis=1)
validation = pd.concat([validation['recommend'], validation.drop(['recommend'], axis=1)], axis=1)
```

```python
import sagemaker
import boto3
bucket = 'sagemaker-studio-s5jna6lkuup'
```

```python
train.to_csv('train.csv', index=False, header=False)
validation.to_csv('validation.csv', index=False, header=False)
```

```python
boto3.Session().resource('s3').Bucket(bucket).Object('train.csv').upload_file('train.csv')
boto3.Session().resource('s3').Bucket(bucket).Object('validation.csv').upload_file('validation.csv')
```

```python
from sagemaker.predictor import csv_serializer
from sagemaker.amazon.amazon_estimator import get_image_uri
container = get_image_uri(boto3.Session().region_name, 'xgboost')
from sagemaker.predictor import csv_serializer
```

```python
s3_input_train = sagemaker.s3_input(s3_data='s3://{}/train.csv'.format(bucket), content_type='csv')
s3_input_validation = sagemaker.s3_input(s3_data='s3://{}/validation.csv'.format(bucket), content_type='csv')
```

```python
from sagemaker import get_execution_role
role = get_execution_role()
sess = sagemaker.Session()
xgb = sagemaker.estimator.Estimator(container,role,train_instance_count=1,train_instance_type='ml.m4.xlarge',output_path='s3://{}/output'.format(bucket),sagemaker_session=sess)
xgb.set_hyperparameters(max_depth=3,eta=0.2,gamma=4,min_child_weight=4,subsample=0.8,silent=0,objective='binary:logistic',num_round=20)
xgb.fit({'train': s3_input_train,'validation': s3_input_validation})
```

```python
xgb_predictor = xgb.deploy(initial_instance_count=1,instance_type='ml.m4.xlarge')
```

```python
xgb_predictor
```

```python
xgb_predictor.content_type = 'text/csv'
xgb_predictor.serializer = csv_serializer
```

```python
import numpy as np
def predict(data, rows=500):
    split_array = np.array_split(data, int(data.shape[0] / float(rows) + 1))
    predictions = ''
    for array in split_array:
        predictions = ','.join([predictions, xgb_predictor.predict(array).decode('utf-8')])
    return np.fromstring(predictions[1:], sep=',')
```

```python
# validating using validation set 
val_label = validation['recommend']
validation = validation.drop(['recommend'], axis=1)
```

```python
validation.to_numpy()[0]
```

```python
pred_val = predict(validation.to_numpy())
```

```python
pred_val
```

```python
pred_val[pred_val >= 0.5] = 1
pred_val[pred_val < 0.5] = 0
pred_val
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(val_label, pred_val)
```
