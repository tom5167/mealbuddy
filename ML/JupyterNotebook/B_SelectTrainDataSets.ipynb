```python
import pandas as pd
```

```python
dataframe=pd.read_csv("yelp_restaurant.csv")
print("dataframe>>")
dataframe
```

```python
dataframe.describe()
```

```python
df=dataframe.fillna(dataframe.mean())
df.describe()
```

```python
rc_avg=df["review_count"].mean()
rc_avg
```

```python
rating_avg=df['rating'].mean()
rating_avg
```

```python
def filterFunc(row):
    if row['rating']>rating_avg and row['review_count']>rc_avg:
        val=1
    else:
        val=0
    return val
```

```python
df['recommend'] = df.apply(filterFunc,axis=1)
df
```

```python
df_recommend = df[df['recommend'] == 1].sort_values(by=['review_count', 'rating'],ascending=[True,False])
df_recommend
```

```python
df_non_recommend=df[df['recommend'] == 0].sort_values(by=['rating','review_count'],ascending=[True,False])
df_non_recommend
```

```python
len(df_recommend)+len(df_non_recommend)
```

```python
#80% from df_recommend
#80% from df_non_recommend
#into semisupervised set

good = 0.8
bad = 0.2
count_rec = len(df_recommend)
count_non_rec = len(df_non_recommend)

#80% of recommend
df1 = df_recommend.tail(int(good*count_rec))

#80% of non_recommend
df2 = df_non_recommend.tail(int(good*count_non_rec))

semisupervised_train = df1.append(df2)
##80% of recommend+20% of non_recommend
semisupervised_train
```

```python
df1 = df_recommend.head(int(bad*count_non_rec))
df2 = df_non_recommend.head(int(bad*count_non_rec))
unsupervised_train = df1.append(df2)
unsupervised_train
```

```python
len(semisupervised_train)+len(unsupervised_train)
```

```python
len(df)-(len(semisupervised_train)+len(unsupervised_train))
```

```python
filename = 'semisupervised_train.csv'
semisupervised_train_csvFile = open(filename, 'a')
semisupervised_train.to_csv(filename, index = False)
```

```python
filename = 'unsupervised_train.csv'
unsupervised_train_csvFile = open(filename, 'a')
unsupervised_train.to_csv(filename, index = False)
```
