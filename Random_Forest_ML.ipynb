```python
dataframe=pd.read_csv("C:\\Users\\ELVIN\\AppData\\Local\\Programs\\Python\\Python38-32\\Scripts\\yelp_restaurant.csv")
print("dataframe>>")
dataframe
```

```python
dataframe_re=dataframe.replace(to_replace="?",value="NaN")
print("dataframe>>")
dataframe
```

```python

df=dataframe.fillna(dataframe.mean())
rc_avg=df["review_count"].mean()
rating_avg=df['rating'].mean()

rc_avg
rating_avg

```

```python
import numpy as np
```

```python
def f(row):
    if row['rating']>rating_avg and row['review_count']>rc_avg:
        val=1
    else:
        val=0
    return val
```

```python
df['recommend'] = df.apply(f,axis=1)
```

```python
df
```

```python
df_recommend = df[df['recommend'] == 1].sort_values(by=['review_count', 'rating'],ascending=[True,False])
```

```python
df_recommend
```

```python
df_non_recommend=df[df['recommend'] == 0].sort_values(by=['rating','review_count'],ascending=[True,False])
```

```python
df_non_recommend
```

```python
#80% from df_recommend
#80% from df_non_recommend
#into semisupervised set

good = 0.8
bad = 0.2
count_rec = len(df_recommend)
count_non_rec = len(df_non_recommend)

#80% of recommend
df1 = df_recommend.tail(int(good*count_rec))

#80% of non_recommend
df2 = df_non_recommend.tail(int(good*count_non_rec))

semisupervised_train = df1.append(df2)
##80% of recommend+20% of non_recommend
semisupervised_train
```

```python

df1 = df_recommend.head(int(bad*count_non_rec))
df2 = df_non_recommend.head(int(bad*count_non_rec))
unsupervised_train = df1.append(df2)
unsupervised_train
```

```python
len(unsupervised_train)+len(semisupervised_train)
```

```python
semisupervised_train_for_model=semisupervised_train.drop(['id', 'cuisine'], axis=1)
semisupervised_train_for_model
```

```python
features_cols=[ 'review_count', 'rating']
```

```python
X = semisupervised_train_for_model[features_cols] # Features
y= semisupervised_train.recommend
```

```python
%pip install  scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=1)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python

y_predicted = model.predict(X_test)
y_predicted
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```
